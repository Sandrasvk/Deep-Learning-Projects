### Create a function to preprocess the text

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
stop_words = set(stopwords.words('english'))
lemma = WordNetLemmatizer()
def preprocess(sentence):
    text = re.sub(r"[^a-zA-Z0-9]", ' ',sentence).split()

    words = [x.lower() for x in text if x not in stop_words]

    word = [lemma.lemmatize(word,'v') for word in words]
    word = ' '.join(word)
    return word
```

```python
### Apply the function on Text column
```

```python
df['review'] = df['review'].apply(preprocess)
```

### Split the dataset into Input and Target Variables

```python
X = df['review']
y = df['sentiment']
```

### One Hot encode Target variables

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
encoder = LabelBinarizer()
```

```python
y = encoder.fit_transform(y)
```

```python
#Check encoder classes
labels = encoder.classes_
labels
```

### Split Dataset into training and testing set

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

### Convert text to Sequence and add padding

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
tokenizer = Tokenizer(num_words=5000)
```

```python
tokenizer.fit_on_texts(X_train)
```

```python
vocab_size = len(tokenizer.word_index)+1
```

```python
def text_encoding(txt):
    encoded_docs = tokenizer.texts_to_sequences(txt)

    padded_sequence = pad_sequences(encoded_docs,maxlen=128)
    return padded_sequence
```

```python
#Applying function on both Training and Testing Set

X_train = text_encoding(X_train)
X_test = text_encoding(X_test)
```

```python
tokenizer
```

```python
vocab_size
```

### Create a keras Sequential model with LSTM Layers

```python
from keras import Sequential
from keras import layers
```

```python
model = Sequential()

model.add(layers.Embedding(vocab_size,128))

model.add(layers.LSTM(128,return_sequences=True))
model.add(layers.Dropout(0.2))

model.add(layers.LSTM(128))
model.add(layers.Dropout(0.2))

model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dropout(0.2))
model.add(layers.Dense(1,activation='sigmoid'))
```

### Compile the model

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
### Train the model
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=128,validation_split=0.2)
```

### Make Prediction with X_test

```python
y_pred = model.predict(X_test)
```

### Evaluate our model

```python
model.evaluate(X_train,y_train)
```

```python
y_pred_binary = (y_pred > 0.5).astype(int)
accuracy_score(y_test, y_pred_binary)
```

### Confusion matrix on heatmap

```python
sns.heatmap(confusion_matrix(y_test,y_pred_binary),annot=True,fmt='d')
plt.show()
```

### Classification Report

```python
print(classification_report(y_test,y_pred_binary))
```
